# Single LL

```python
#Creating Node
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

#Creating Link Between Nodes
class SLL:
    def __init__(self):
        self.head=None
        
    #To display
    def display(self):
        if self.head is None:
            print("EMPTY!")
        else:
            temp=self.head
        while temp:
            print(temp.data,"--> ",end="")
            temp=temp.next
        
    #insert at begining
    def insert_begin(self,data):
        nb=Node(data)
        nb.next=self.head
        self.head=nb
        
    #insert at end
    def insert_end(self,data):
        ne=Node(data)
        temp=self.head
        while temp.next:
            temp=temp.next
        temp.next=ne
        
    #insert at specified position
    def insert_sp(self,pos,data):
        np=Node(data)
        temp=self.head
        for i in range(pos-1):
            temp=temp.next
        np.next=temp.next
        temp.next=np
        
    #Delete at begining
    def delete_begin(self):
        temp=self.head
        self.head=temp.next
        temp.next=None 
        
    #Delete at end
    def delete_end(self):
        prev=self.head
        temp=self.head.next
        while temp.next is not None:
            temp=temp.next
            prev=prev.next
        prev.next=None
        
    #Delete at specified position
    def delete_sp(self,pos):
        prev=self.head
        temp=self.head.next
        for i in range(pos-1):
            temp=temp.next
            prev=prev.next
        prev.next=temp.next
        temp.next = None
    

        
#To access Nodes create object
L=SLL()
n1=Node(10)
L.head=n1
n2=Node(20)
n1.next=n2
n3=Node(30)
n2.next=n3
#creation of Nodes
print("Nodes after creation: ")
#displaying
L.display()
print()

#inserting value at begining
print("\nAfter inserting node at begining: ")
L.insert_begin(5) #5 is inserted at begining
#displaying
L.display()
print()

#inserting value at ending
print("\nAfter inserting node at ending: ")
L.insert_end(40) #40 is inserted at ending
#displaying
L.display()
print()

#inserting value at specified position
print("\nAfter inserting node at specific position: ")
L.insert_sp(2,15) #15 at 2 position
#displaying
L.display()
print()


#delete node at beginning
print("\nAfter deleting node at beginning: ")
L.delete_begin() 
#displaying
L.display()
print()

#delete node at ending
print("\nAfter deleting node at ending: ")
L.delete_end()
#displaying
L.display()
print()

#delete node at specific position
print("\nAfter deleting node at specific position: ")
L.delete_sp(2)
#displaying
L.display()
print()

# #searching
# L.search()
```

```python
#Creating Node

class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

#Creating Link Between Nodes

class SLL:
    def __init__(self):
        self.head=None
        
#insert at begining

    def insert_begin(self,data):
        nb=Node(data)
        nb.next=self.head
        self.head=nb
        
#insert at end

    def insert_end(self,data):
        ne=Node(data)
        temp=self.head
        while temp.next:
            temp=temp.next
        temp.next=ne

#insert at specified position

    def insert_sp(self,pos,data):
        np=Node(data)
        temp=self.head
        for i in range(pos-1):
            temp=temp.next
        np.data=data
        np.next=temp.next
        temp.next=np

#Delete at begining

    def delete_begin(self):
        temp=self.head
        self.head=temp.next
        temp.next=None
        
#Delete at end

    def delete_end(self):
        prev=self.head
        temp=self.head.next
        while temp.next is not None:
            temp=temp.next
            prev=prev.next
        prev.next=None

#Delete at specified position

    def delete_sp(self,pos):
        prev=self.head
        temp=self.head.next
        for i in range(1,pos-1):
            temp=temp.next
            prev=prev.next
        prev.next=temp.next
        
#searching an element
    
    def search(self):
        x = int(input("\nEnter an element to be serched in LL: "))
        temp = self.head
        pos = 0
        flag = False
        
        #checks whether list is empty
        if self.head == None:
            print("Empty")
        else:
            while temp != None:
            
            #Compares node to be found with each node present in list
                if x == temp.data:
                    flag = True
                    break
                pos += 1    
                temp = temp.next
    
            if flag == True:
                print(x," is found in the list at position ",pos)
            else:
                print(x," is not found")
                
        
#To display
    def display(self):
        if self.head is None:
            print("EMPTY!")
        else:
            temp=self.head
        while temp:
            print(temp.data,"--> ",end="")
            temp=temp.next
    
        
#To access Nodes create object
L=SLL()
n1=Node(10)
L.head=n1
n2=Node(20)
n1.next=n2
n3=Node(30)
n2.next=n3

#inserting value at begining
L.insert_begin(5) #5 is inserted at begining
L.insert_begin(3) #3 is inserted at begining

#inserting value at end
L.insert_end(40) #40 is inserted at end
L.insert_end(45) #45 is inserted at end

#inserting value at specified Position
L.insert_sp(4,35)  #35 is inserted at position 4

#Deleting value at begining 
L.delete_begin()  #3 is removed at begining


#Deleting value at end
L.delete_end()  #45 is removed at end


#Deleting value at specified Position
L.delete_sp(4) #removing element at position 4 i.e;35

#Displaying
L.display()

#searching
L.search()
```
